```python
import matplotlib.pyplot as plt
import gymnasium as gym
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False, render_mode="rgb_array")
```

```python
state, info = env.reset()
```

```python
state
```

```python
frame = env.render()
plt.imshow(frame)
plt.axis("off")
plt.show()
```

### アクションとして、`env`からサンプリングする: `.action_space.sample()`

```python
action = env.action_space.sample()
action
```

```python
next_state, reward, done, truncated, info = env.step(action)

print(f"{next_state=}")
print(f"{reward=}")
print(f"{done=}")
# ループして立ち止まってないか？途中で打ち切り設定ができ、今の状態を確認できる
print(f"{truncated=}")
# そのアクションがどの確率で選択されたか？
print(f"{info=}")
```
